```python
import pandas as pd
from datetime import datetime
import locale
import pandas as pd
from pyproj import Proj, transform
```

```python
df = pd.read_excel('Aguas_Barcelona_saturacionO2_plaguicidas.xlsx')
```

```python
df
```

```python


# Definir el sistema de referencia de coordenadas UTM (suponiendo UTM Zona 33N)
utm_proj = Proj(proj='utm', zone=31, datum='WGS84')

# Definir el sistema de referencia de coordenadas geográficas (Lat, Lon)
wgs84_proj = Proj(proj='latlong', datum='WGS84')

# Función para convertir las coordenadas UTM a latitud y longitud
def utm_to_latlon(utm_x, utm_y):
    lon, lat = transform(utm_proj, wgs84_proj, utm_x, utm_y)
    return lat, lon

# Aplicar la conversión a las columnas del DataFrame
df[['Latitud', 'Longitud']] = df.apply(lambda row: pd.Series(utm_to_latlon(row['UTM X'], row['UTM Y'])), axis=1)

print(df)

```

```python
df
```

```python
# In te column 'Valor', we remove non numerical characters, except for '.'
df['Valor'] = df['Valor'].str.replace('[^0-9.]', '', regex=True)
```

```python
df
```

```python
import pandas as pd

# Iterar sobre los valores únicos de la columna 'Variable'
for value in df['Variable'].unique():
    # Crear un DataFrame para cada valor único de 'Variable'
    sub_df = df[df['Variable'] == value]
    
    # Crear un nombre dinámico para el archivo CSV (usando el valor de 'Variable')
    file_name = f"df_{value}.xlsx"
    
    # Exportar cada sub-DataFrame a un archivo CSV
    sub_df.to_excel(file_name, index=False)
    print(f"Se ha creado el archivo {file_name}")
```

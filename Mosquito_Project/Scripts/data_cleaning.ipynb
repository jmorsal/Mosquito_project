```python
import pandas as pd
```

# AEMET dataset

```python
aemet_port_df = pd.read_csv('..\Datos\AEMET_Port_2124.csv')
aemet_Drassanes_df = pd.read_csv('..\Datos\AEMET_Drassanes_2124.csv')
```

```python
aemet_Drassanes_df
```

```python
# The Port station includes wind data but was inactive for 10 months. However, the Drassanes station is very close, so we can use it to fill the missing data
aemet_port_df['fecha'] = pd.to_datetime(aemet_port_df['fecha'])
aemet_Drassanes_df['fecha'] = pd.to_datetime(aemet_Drassanes_df['fecha'])

# We filter the dates that are in aemet_Drassanes_df but not in aemet_port_df and concatenate them
missing_dates = aemet_Drassanes_df[~aemet_Drassanes_df['fecha'].isin(aemet_port_df['fecha'])]

aemet_df = pd.concat([aemet_port_df, missing_dates], ignore_index=True)
```

```python
aemet_df
```

```python
# Before anything else we have to correct the format of the floats
aemet_df = aemet_df.apply(lambda x: x.str.replace(',', '.', regex=False) if x.dtype == 'object' else x)
# We convert the columns to numeric values
aemet_df.iloc[:, 4:16] = aemet_df.iloc[:, 4:16].apply(pd.to_numeric, errors='coerce')
```

```python
aemet_df
```

NaN managing

```python
aemet_df.isna().sum()
```

```python
# The wind values are missing for a period too long, so we should not fill them
fill_NaN_cols = ['tmed', 'tmin', 'prec','tmax', 'hrMedia', 'hrMax', 'hrMin']
```

```python
aemet_df.columns
```

```python
# Delete unnecesary columns
aemet_df = aemet_df.drop(['indicativo', 'nombre', 'provincia', 'altitud', 'horatmin', 'horatmax', 'horaHrMax', 'horaHrMin', 'horaracha'], axis=1)
```

```python
# We create a function to fill the NaN values with the mean of the previous and next values
def fill_with_mean(df, columns):
    for column in columns:
        # Iterate over the rows to identify and fill the NaNs in each column
        for i in range(1, len(df) - 1): 
            if pd.isna(df[column][i]):
                prev_value = df[column][i - 1] if pd.notna(df[column][i - 1]) else None
                next_value = df[column][i + 1] if pd.notna(df[column][i + 1]) else None

                # If both before and after values ​​are available, calculate the average
                if prev_value is not None and next_value is not None:
                    # Calculate the average of the available values
                    df[column][i] = (prev_value + next_value) / 2
                elif prev_value is not None:  # If only the previous value is available
                    df[column][i] = prev_value
                elif next_value is not None:  # If only the posterior value is available
                    df[column][i] = next_value

        # If there is a full block of NaNs at the end, you can fill them with the last available value
        if pd.isna(df[column].iloc[-1]): 
            last_non_na = df[column].dropna().iloc[-1]
            df[column].iloc[-1] = last_non_na

    return df



aemet_df = fill_with_mean(aemet_df, fill_NaN_cols)
```

# Mosquito dataset

```python
mosquito_df = pd.read_excel('..\Datos\observations.xlsx')
```

```python
mosquito_df
```

```python
# We group the data by day and count the number of observations
mosquito_df['date'] = pd.to_datetime(mosquito_df['date'])

mosquito_df['fecha_solo_dia'] = mosquito_df['date'].dt.date

mosquito_simple_df = mosquito_df.groupby('fecha_solo_dia').agg('count')

mosquito_simple_df = mosquito_simple_df.reset_index()

mosquito_simple_df['fecha_solo_dia'] = pd.to_datetime(mosquito_simple_df['fecha_solo_dia'])
```

```python
mosquito_simple_df
```

```python
# We frop all coumns but date and id. Rename id to count and fecha_solo_dia to fecha
mosquito_simple_df = mosquito_simple_df[['id', 'fecha_solo_dia']]
mosquito_simple_df = mosquito_simple_df.rename(columns={'id': 'count', 'fecha_solo_dia': 'fecha'})
mosquito_simple_df
```

# Final dataframe

```python
# We merge the two dataframes
final_df = aemet_df.merge(mosquito_simple_df, how='outer', on='fecha')
final_df
```

```python
# When there is no mosquito we have to add 0 to the count column
final_df['count'].fillna(0, inplace=True)
```

```python
# We need to add a column with 1 if there is any number in count and 0 if not
final_df['mosquito'] = final_df['count'].apply(lambda x: 1 if x > 0 else 0)
```

```python
# The month may be an important variable, so we add it
final_df['fecha'] = pd.to_datetime(final_df['fecha'])
final_df['dayofyear'] = final_df['fecha'].dt.dayofyear
```
